```python
cast_movie_by_id = cast_df.rename(columns={'id': 'person_id'}).drop('birth', axis=1).join(cast_and_movies_df.set_index('person_id'), on='person_id')

cast_movie_by_id = cast_movie_by_id.join(movies_df.rename(columns={'id': 'movie_id'}).set_index('movie_id'), on='movie_id')
```

```python
def check_id(person_id, ref_id, cast_movie_by_id, initial_count=0):
    count = initial_count
    
    movie_ids_someone = cast_movie_by_id[cast_movie_by_id['person_id'] == person_id]['movie_id']
    person_ids = cast_movie_by_id.set_index('movie_id').loc[movie_ids_someone]['person_id'].unique()
    
    if ref_id in person_ids:
        return 1
    else:
        count += 1
        check_id()
```

```python
check_id(1, 102, cast_movie_by_id)
```

```python
x = pd.Series(cast_movie_by_id.set_index('movie_id').loc[movie_ids_someone]['person_id'].unique())
```

```python
other_ids = x[x!=1]
```

```python
for i in other_ids:
    print(check_id(i, 102, cast_movie_by_id))
```
